```python
import pandas as pd
from scipy import stats
import seaborn as sns
auto = pd.read_csv('auto.csv')
auto.head(10)
```

```python
auto.describe()
```

```python
sns.pairplot(auto, vars=['engine-size', 'compression-ratio', 'horsepower','peak-rpm','city-mpg','price'],
                 kind='scatter');
```

```python
ax = auto.price.hist(bins = 15) 
ax.set_title('Price distribution')
ax.set_xlabel('Price')
ax.set_ylabel('Quantity')
```

```python
skewness=stats.skew(auto.price)
skewness
```

```python
kurtosis=stats.kurtosis(auto.price)
kurtosis
```

```python
corr = auto.corr()
corr
```

```python
stats.ttest_rel(auto['wheel-base'], auto['width'])  
```

```python
atmo_wb= auto[auto['aspiration'] == 'std']['wheel-base']
turbo_wb= auto[auto['aspiration'] == 'turbo']['wheel-base']

stats.ttest_ind(atmo_wb, turbo_wb) 
```
